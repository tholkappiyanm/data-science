# About Dataset
Dataset containing the list of 2500+ people with fortunes valued at least 1 Billion USD.

Dataset Features
1. Rank - Ranking of the Billionare
2. Name - Name of the Billionare
3. Net Worth - Their Net Worth in Billions USD
4. Age - The Age of the Billionare
5. Country - The Country in which the Billionare lives
6. Source - Income Source of the Billionare
7. Industry - Sector/Industry/Market Segment Each Billionaires Have Made their Fortune

This dataset has been obtained from [Kaggle](https://www.kaggle.com/datasets/jjdaguirre/forbes-billionaires-2022).

```python
# Importing all needed libraries
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns

plt.style.use("fivethirtyeight")


print("All Modules & Settings Initialized 🚀")
```

```python
# Import the data
from urllib.request import urlretrieve 
URL = "https://raw.githubusercontent.com/muhammadanas0716/Data-Science-Projects---EDA/main/4.%20Forbes%20Richest%20Poeple%20-%20Analysis/2022_forbes_billionaires.csv"
forbes_data = pd.read_csv(URL)

print("Data Imported Successfully 🚀")
```

```python
# Age distribution among the data
from matplotlib.pyplot import figure
figure(num=None, figsize=(25, 20), dpi=200, facecolor='w', edgecolor='k')

forbes_data["age"].value_counts(ascending=True).plot.barh(color=["#cdb4db", "#ffc8dd", "#ffafcc"])

# Description
plt.title("Age X Frequency")
plt.ylabel("Age")
plt.xlabel("Frequency")

# Save the Figure
plt.savefig('age_distribution.png')
plt.show()
```

```python
# Networth X Frequency
from matplotlib.pyplot import figure
figure(num=None, figsize=(80, 60), dpi=200, facecolor='w', edgecolor='k')

forbes_data["networth"].value_counts(ascending=True).plot.barh(color=["#ffcdb2", "#ffb4a2", "#e5989b"])

# Description
plt.title("Networth X Frequency")
plt.ylabel("Networth")
plt.xlabel("Frequency")

# Save the Figure
plt.savefig('networth_frequency.png')
plt.show()
```

```python
# Industry X Frequency
from matplotlib.pyplot import figure
figure(num=None, figsize=(50, 30), dpi=200, facecolor='w', edgecolor='k')

forbes_data["industry"].value_counts(ascending=True).plot.barh(color=["#2b2d42", "#8d99ae", "#ef233c", "#d90429"])

# Description
plt.title("Industry X Frequency")
plt.ylabel("Industry")
plt.xlabel("Frequency")

# Save the Figure
plt.savefig('industry_distribution.png')
plt.show()
```

```python
# How does Networth Change with age and industry
sns.catplot(data=forbes_data, x="age", y="industry", kind="bar")
```
